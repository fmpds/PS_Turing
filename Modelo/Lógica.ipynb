#Q1 item A

```python
QNT_EMAILS = int(input())
```

```python
emails_list = []
for i in range(QNT_EMAILS):
  emails_list.append(input())
```

```python
# recolhendo o tamanho dos emails
TMNH_EMAILS = len(emails_list[0])
```

```python
# contador de caracteres iguais
qnt_carac = 0
for i in range(1, QNT_EMAILS):
  for j in range(TMNH_EMAILS):

    # verificando se o caracter no email i - 1 (i começando do 1), na posição j é igual ao 
    # caracter j no email i 
    if emails_list[i - 1][j] == emails_list[i][j]:
      qnt_carac += 1
    else:
    # se não for igual, sai do laco interno e volta para o externo
      break; 
```

```python
print(qnt_carac)
```

#Q1 item B

```python
QNT_EMAILS = int(input())
```

```python
emails_list = []
for i in range(QNT_EMAILS):
  emails_list.append(input())
```

```python
for i in range(QNT_EMAILS):
  # recolhendo meio dos emails
  MET_EMAIL = int(len(emails_list[i]) / 2)

  ''' 
  recolhendo substrings dos emails de acordo com a metade do seu tamanho. 
  '''

  # :MET_EMAIL -> 0 até valor da metade
  met_email_1 = emails_list[i][:MET_EMAIL]

  # MET_EMAIL -> valor da metade ate o final
  met_email_2 = emails_list[i][MET_EMAIL:]

  # invertendo as strings com o método ::-1
  met_email_1 = met_email_1[::-1] 
  met_email_2 = met_email_2[::-1]
  
  # contacetando metades no email completo
  email_completo = met_email_1 + met_email_2
  

  # verificando se o @usp.br está correto -> sempre será os ultimas 7 caracteres
  if email_completo[-7:] != '@usp.br':
    print("ERROR")
  else:
    print(email_completo)
```

```python
aux = input()
```

```python
aux = aux.split(sep = ' ')
```

```python
VAGAS = int(aux[0])
INSTANTE = int(aux[1])
```

```python
momento = []
for i in range(INSTANTE):
  momento.append(input())
```

```python
output = []
estacionado = []
for i in momento:

  # verificando se sera realizado a inserção ou remocao do veiculo na vaga
  # e se existe espaco na garagem 
  if int(i) > 0 and len(estacionado) < VAGAS:
    estacionado.append(i)
  else:
    # realizando remocao dos carros das vagas
    # verificando se o ultimo carro estacionado eh o que esta sendo retirado
    if int(estacionado[-1]) == abs(int(i)):
      estacionado.pop()
    else:
      output.append("não")
      break
  
  output.append("sim")
```
